```python
def reverse(n):
    a = 0
    while n > 0:
        i = n % 10
        a = a*10 + i
        n //= 10
    return a

def Palidrone(n):
    return n == reverse(n)

n = int(input())
steps = 0
while not Palidrone(n):
    n += reverse(n)
    steps += 1
    if steps > 100: break
    
print(f"Số bước thực hiện: {steps}")
print(f"Giá trị palindrome thu được: {n}")
    
```

```python
n  = int(input(""))
steps = 0

while True:
    s = str(n)
    flag = True
    for i in range(len(s)//2):
        if s[i] != s[len(s)-1-i]:
            flag = False
            break
    rev = 0
    temp = n
    while temp > 0:
        rev = rev * 10 + temp % 10
        temp //= 10
    if rev == n:
        break
    n += rev
    steps += 1
    if steps > 100:
        break

print(steps)
print(n)
```
