```python
#W1A1
# Gán (lưu) giá trị số 7 vào biến có tên là 'a'
a = 7

# Gán giá trị số 5 vào biến có tên là 'b'
b = 5

# Thực hiện phép trừ: lấy giá trị của 'a' trừ đi giá trị của 'b'
# và lưu kết quả vào một biến mới tên là 'c'
c = a - b

# In (hiển thị) giá trị hiện tại đang được lưu trong biến 'c' ra màn hình
print(c)
```

```python
#W1A2
# Gán chuỗi ký tự "Hà Nội" cho biến có tên là 'city'
city = "Hà Nội"

# Gán giá trị số nguyên 2025 cho biến có tên là 'year'
year = 2025

# In ra màn hình một chuỗi được ghép từ nhiều phần.
# Hàm print() sẽ tự động chèn một khoảng trắng vào giữa các đối số được ngăn cách bởi dấu phẩy.
print("Thành phố:", city, "- Năm:", year)
```

```python
#W1A3
# Gán giá trị 4 cho biến n, n là giới hạn trên của dãy số cần tính tổng
n = 4

# Khởi tạo biến t (total - tổng) với giá trị ban đầu là 0
t = 0

# Bắt đầu vòng lặp for. Biến i sẽ lần lượt nhận các giá trị từ 1 đến n (tức là 1, 2, 3, 4)
# range(1, n + 1) sẽ tạo ra dãy số [1, 2, 3, 4]
for i in range(1, n + 1):
  
  # Cộng giá trị hiện tại của i vào biến tổng t. Dòng này tương đương với "t = t + i"
  t += i

# Sau khi vòng lặp kết thúc, in giá trị cuối cùng của biến t ra màn hình
print(t)
```

```python
#W1A4

# Tạo một danh sách (list) chứa các số nguyên và gán vào biến 'numbers'
numbers = [1, 2, 3, 4]

# Bắt đầu vòng lặp 'for', duyệt qua từng phần tử trong danh sách 'numbers'.
# Ở mỗi lần lặp, phần tử hiện tại sẽ được gán cho biến 'x'.
for x in numbers:
  
  # Kiểm tra xem phần dư của phép chia 'x' cho 2 có bằng 0 không.
  # Nếu bằng 0, x là số chẵn.
  if x % 2 == 0:
    
    # In ra màn hình thông báo x là số chẵn.
    print(x, "là số chẵn")
    
  # Nếu điều kiện ở trên (if) là sai.
  else:
    
    # In ra màn hình thông báo x là số lẻ.
    print(x, "là số lẻ")
```

```python
#w1A5

# Tạo một danh sách (list) chứa các chuỗi và gán vào biến 'animals'
animals = ["cat", "dog", "cat", "bird"]

# Khởi tạo một biến đếm với giá trị ban đầu là 0
count = 0

# Bắt đầu vòng lặp 'for', duyệt qua từng phần tử trong danh sách 'animals'.
# Biến 'a' sẽ tạm thời giữ giá trị của phần tử hiện tại (dù không được sử dụng trực tiếp).
for a in animals:
  
  # Ở mỗi lần lặp, tăng giá trị của biến đếm 'count' lên 1 đơn vị.
  count += 1

# Sau khi vòng lặp kết thúc, in ra chuỗi thông báo và giá trị cuối cùng của biến 'count'.
print("Số phần tử trong danh sách là:", count)
```

```python
#W1A7

# LỖI 1: Dữ liệu nhập vào từ input() là dạng chuỗi (string), cần chuyển sang dạng số nguyên (integer) bằng hàm int()
num = int(input("Nhập số: ")) 

# LỖI 2: Phép so sánh bằng trong Python dùng hai dấu bằng "==", một dấu "=" là phép gán giá trị
if num % 2 == 0: 
  print("Số chẵn") 
  
# LỖI 3: Câu lệnh "else" phải kết thúc bằng dấu hai chấm ":"
else: 
  print("Số lẻ")
```

```python
#W1A8

# Bắt đầu vòng lặp for. range(3) sẽ tạo ra một dãy số (0, 1, 2), nên vòng lặp sẽ chạy 3 lần.
for i in range(3): 
  
  # Dòng này nằm BÊN TRONG vòng lặp. Nó sẽ in ra số lần lặp.
  # Ta dùng "i + 1" để hiển thị kết quả từ 1, 2, 3 thay vì 0, 1, 2.
  print("AI đang học lần", i + 1) 

# Dòng này nằm BÊN NGOÀI vòng lặp và chỉ được thực thi 1 lần sau khi vòng lặp kết thúc.
print("Huấn luyện xong!")
```

```python
#W1A9

# Bắt đầu vòng lặp for, duyệt qua từng phần tử trong danh sách ["cat", "dog", "fish"].
# Ở mỗi lần lặp, phần tử hiện tại sẽ được gán cho biến x.
for x in ["cat", "dog", "fish"]: 
  
  # In ra chuỗi "Dự đoán con vật:" theo sau là giá trị hiện tại của biến x.
  print("Dự đoán con vật:", x)
```
