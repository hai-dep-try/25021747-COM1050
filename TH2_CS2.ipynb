```python
#W2A3
a, b = map(int, input().split())
print(f'{a} + {b} = {a+b}')
print(f'{a} - {b} = {a-b}')
print(f'{a} / {b} = {a/b :.2f}')
print(f'{a} * {b} = {a*b}')
print(f'{a} // {b} = {a//b}')
print(f'{a} % {b} = {a%b}')
```
