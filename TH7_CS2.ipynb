```python
#W7A1
def binary_search(arr, left, right, target):
    if left > right: 
        return -1
    mid = (left + right)//2
    if arr[mid] > target:
        return binary_search(arr, left, mid - 1, target)
    elif arr[mid] < target:
        return binary_search(arr, mid + 1, right, target)
    else:
        return mid
arr = list(map(int, input().split()))
target = int(input())
print(binary_search(arr, 0, len(arr) - 1, target))
```

```python
#W7A3
def BubbleSort(a):
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a

def SelectionSort(a):
    n = len(a)
    for i in range(n):
        min = i
        for j in range(i + 1, n):
            if a[min] > a[j]:
                min = j
            if min != i:
                a[min], a[i] = a[i], a[min]
    return a

arr = list(map(int, input().split()))
print(BubbleSort(arr))
print(SelectionSort(arr))
```

```python
#W7A7
def daycontang(a):
    n = len(a)
    if n == 0: 
        return 0
    dp = [1] * n

    for i in range(n):
        for j in range(i):
            if a[i] > a[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)

a = eval(input())
print(daycontang(a))
```

```python
#W7A7 bisect
import bisect
def daycontang(a):
    n = len(a)
    if n == 0: 
        return 0
    sub = []
    for x in a:
        if not sub or x > sub[-1]:
            sub.append(x)
        else:
            idx = bisect.bisect_left(sub, x)
            sub[idx] = x
    return len(sub)

a = eval(input())
print(daycontang(a))
```

```python
#W7A8
#intervals = eval(input())
#queries = eval(input())

import heapq  # Thư viện để dùng Hàng đợi ưu tiên (Min-Heap)

def min_interval_length(intervals, queries):
    # --- BƯỚC 1: SẮP XẾP INTERVALS ---
    # Sắp xếp các khoảng dựa trên điểm bắt đầu (Start)
    # intervals[i][0] là start, intervals[i][1] là end
    intervals.sort()

    # --- BƯỚC 2: CHUẨN BỊ QUERIES ---
    # Vì mình cần sắp xếp queries để quét, nhưng lại phải trả về kết quả
    # đúng thứ tự ban đầu, nên em sẽ lưu thêm chỉ số index gốc vào.
    queries_with_index = []
    
    # Dùng vòng lặp for cơ bản thay vì cú pháp tắt
    for i in range(len(queries)):
        q_val = queries[i]
        # Lưu thành cặp [giá trị query, vị trí gốc]
        queries_with_index.append([q_val, i])
        print(queries_with_index)
    
    # Sắp xếp queries tăng dần theo giá trị
    queries_with_index.sort()

    # --- BƯỚC 3: XỬ LÝ CHÍNH (LINE SWEEP) ---
    min_heap = []   # Heap chứa các khoảng [độ dài, điểm kết thúc]
    
    # Tạo mảng kết quả, mặc định điền toàn bộ là -1
    result = [-1] * len(queries)
    
    interval_idx = 0  # Con trỏ để duyệt danh sách intervals
    n = len(intervals)

    # Duyệt qua từng query đã sắp xếp
    for i in range(len(queries_with_index)):
        current_query_val = queries_with_index[i][0]
        original_index = queries_with_index[i][1]

        # 3.1: Nạp các Interval hợp lệ vào Heap
        # Điều kiện: Interval phải bắt đầu trước hoặc ngay tại vị trí query (Start <= Query)
        while interval_idx < n and intervals[interval_idx][0] <= current_query_val:
            start = intervals[interval_idx][0]
            end = intervals[interval_idx][1]
            
            # Tính độ dài (theo công thức: cuối - đầu + 1)
            length = end - start + 1
            
            # Đẩy vào Heap: ưu tiên độ dài nhỏ nhất lên đầu
            # Cấu trúc phần tử trong heap: [độ dài, điểm kết thúc]
            heapq.heappush(min_heap, [length, end])
            print(min_heap)
            
            # Tăng con trỏ để xét interval tiếp theo
            interval_idx = interval_idx + 1

        # 3.2: Dọn dẹp Heap (Loại bỏ các Interval đã hết hạn)
        # Điều kiện: Interval kết thúc trước vị trí query (End < Query) thì không che được nữa
        while len(min_heap) > 0 and min_heap[0][1] < current_query_val:
            heapq.heappop(min_heap)
            print(min_heap) # Vứt interval ở đỉnh heap đi

        # 3.3: Lấy kết quả
        # Nếu Heap còn phần tử, thì phần tử đỉnh (nhỏ nhất) chính là đáp án
        if len(min_heap) > 0:
            smallest_len = min_heap[0][0]
            result[original_index] = smallest_len
        print(result)
            
    return result

# --- CHẠY THỬ VỚI VÍ DỤ ---
# Ví dụ 1 từ đề bài
intervals = [[1, 4], [2, 4], [3, 6], [4, 4]]
queries = [2, 3, 4, 5]

print("Kết quả:", min_interval_length(intervals, queries))
    

    
```

```python
def min_interval_length(intervals, queries):
    results = []

    for i in range(len(queries)):
        q_val = queries[i]

        current_min_len = -1
        for j in range(len(intervals)):
            start = intervals[j][0]
            end = intervals[j][1]
            
            if start <= q_val and q_val <= end:
                length = end - start + 1

                if current_min_len == -1 or length < current_min_len:
                    current_min_len = length
        results.append(current_min_len)
    
    return results
intervals = [[1, 4], [2, 4], [3, 6], [4, 4]]
queries = [2, 3, 4, 5]

print("Kết quả (Trâu):", min_interval_length(intervals, queries))

```
