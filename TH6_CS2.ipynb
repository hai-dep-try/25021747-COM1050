```python
#W6A9
def nhap_hai_dictionary():
    """
    Nhập vào hai dictionary từ người dùng.
    """
    print("Nhập dictionary thứ nhất (ví dụ: {'key1': 'value1', 'key2': 2}):")
    dict1_str = input()
    dict1 = eval(dict1_str)

    print("Nhập dictionary thứ hai (ví dụ: {'key3': 'value3', 'key4': 4}):")
    dict2_str = input()
    dict2 = eval(dict2_str)
    dict3 = {}
    for i in dict2:
        dict3[i] = dict2[i]
        
    for i in dict1:
        if dict1[i] not in dict3:
            dict3[i] = dict1[i]
        else:
            dict3[i] += dict1[i]
            
    return dict3
dict3 = nhap_hai_dictionary()
print(dict3)
```

```python
def encrypt_number(number):
    s = str(number)
    digit = []
    for i in range(len(s)):
        number = (int(s[i]) + 7) % 10

        digit.append(number)
    digit[0], digit[2] = digit[2], digit[0]
    digit[1], digit[3] = digit[3], digit[1]
    result = "".join(str(d) for d in digit)
    return result

n = int(input())
code = encrypt_number(n)
print(code)
```

```python
def create_diamond(K, border_char, inner_char, stitch_char):
    """
    Hàm in ra hình kim cương rỗng có kích thước K.
    - K: Số hàng của nửa trên (bao gồm cả hàng giữa).
    - border_char: Ký tự viền.
    - inner_char: Ký tự điền bên trong.
    - stitch_char: Ký tự đặc biệt ở viền hàng giữa.
    """
    
    # Tổng số hàng của hình kim cương là 2K - 1
    total_rows = 2 * K - 1
    
    # Duyệt qua từng hàng (index từ 0 đến total_rows - 1)
    for i in range(total_rows):
        # Tính khoảng cách từ hàng hiện tại đến hàng giữa (hàng K-1)
        # Hàng giữa có index là K-1.
        # Ví dụ K=5 (tổng 9 hàng), hàng giữa là index 4.
        dist = abs(K - 1 - i)
        
        # Số khoảng trắng (padding) bên ngoài bên trái = dist
        padding = " " * dist
        
        # Xử lý Hàng Giữa (Middle Row)
        if i == K - 1:
            # Ở hàng giữa, viền là stitch_char
            current_border = stitch_char
        else:
            # Các hàng khác, viền là border_char
            current_border = border_char
            
        # Xử lý việc in
        if dist == K - 1:
            # Đây là đỉnh trên cùng hoặc đỉnh dưới cùng (chỉ có 1 ký tự)
            # Trừ trường hợp đặc biệt K=1 thì đỉnh cũng là giữa
            if K == 1:
                print(stitch_char)
            else:
                print(padding + border_char)
        else:
            # Các hàng ở giữa (có 2 viền và phần inner bên trong)
            # Tính độ rộng phần bên trong
            # Công thức: Tổng độ rộng tại hàng i là (2K-1) - 2*dist
            # Trừ đi 2 ký tự viền => số lượng inner_char = (2K-1) - 2*dist - 2
            # Rút gọn: 2*(K - 1 - dist) - 1
            inner_width = 2 * (K - 1 - dist) - 1
            inner_part = inner_char * inner_width
            
            # In ra: Padding + Viền trái + Nội dung trong + Viền phải
            print(padding + current_border + inner_part + current_border)

# --- CHẠY THỬ (TEST CASES) ---
if __name__ == "__main__":
    print("--- Test Case 1: K=5, border='@', inner='.', stitch='S' ---")
    create_diamond(5, '@', '.', 'S')
    
    print("\n--- Test Case 2: K=3, border='#', inner=' ', stitch='X' ---")
    create_diamond(3, '#', ' ', 'X')
    
    print("\n--- Test Case 3: K=1 (Trường hợp đặc biệt) ---")
    create_diamond(1, '*', '-', 'O')
```

```python
def create_diamond(K, border_char, inner_char, stitch_char):
    total_rows = 2 * K - 1
    for i in range(total_rows):
        index = abs(K - 1 - i)
        padding = " " * index
        if i == K - 1:
            a = stitch_char
        else:
            a = border_char
        tmp = 2*(K - 1) - 1 - index*2
        if index == K - 1: print(padding + a)
        else: print(padding + a + inner_char*tmp + a)

if __name__ == "__main__":
    print("--- Test Case 1: K=5, border='@', inner='.', stitch='S' ---")
    create_diamond(5, '@', '.', 'S')
    
    print("\n--- Test Case 2: K=3, border='#', inner=' ', stitch='X' ---")
    create_diamond(3, '#', ' ', 'X')
    
    print("\n--- Test Case 3: K=1 (Trường hợp đặc biệt) ---")
    create_diamond(1, '*', '-', 'O')
```
