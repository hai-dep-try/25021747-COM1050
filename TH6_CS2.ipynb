```python
#W6A4
a1 = list(map(int, input().split()))
dic = {"postive" : 0, "negative" : 0, "zeros" : 0}
for i in range(len(a1)):
    if a1[i] > 0:
        dic["postive"] += 1
    elif a1[i] < 0:
        dic["negative"] += 1
    else:
        dic["zeros"] += 1
print(dic)
```

```python
#W6A9
def nhap_hai_dictionary():
    """
    Nhập vào hai dictionary từ người dùng.
    """
    print("Nhập dictionary thứ nhất (ví dụ: {'key1': 'value1', 'key2': 2}):")
    dict1_str = input()
    dict1 = eval(dict1_str)

    print("Nhập dictionary thứ hai (ví dụ: {'key3': 'value3', 'key4': 4}):")
    dict2_str = input()
    dict2 = eval(dict2_str)
    dict3 = {}
    for i in dict2:
        dict3[i] = dict2[i]
        
    for i in dict1:
        if dict1[i] not in dict3:
            dict3[i] = dict1[i]
        else:
            dict3[i] += dict1[i]
            
    return dict3
dict3 = nhap_hai_dictionary()
print(dict3)
```
