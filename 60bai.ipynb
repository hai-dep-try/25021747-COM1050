```python
#bai57
# Tên file để lưu trữ
ten_file = 'hieulo.txt'

# --- PHẦN 1: NHẬP VÀ GHI FILE ---
print(f"--- Bắt đầu nhập số vào file '{ten_file}' ---")
print("Nhập một số nguyên, sau đó nhấn Enter.")
print("(Nhấn Enter lần nữa mà không nhập gì để kết thúc)")

try:
    # Mở file với chế độ 'w' (write - ghi)
    # 'w' sẽ tạo file mới (hoặc ghi đè file cũ nếu đã tồn tại)
    with open(ten_file, 'w', encoding='utf-8') as f:
        while True:
            # 1. Yêu cầu người dùng nhập
            chuoi_nhap = input("Nhập số: ")
            
            # 2. Kiểm tra nếu người dùng nhấn Enter (chuỗi rỗng) thì dừng lại
            if chuoi_nhap == "":
                break
                
            # 3. Thử chuyển đổi chuỗi nhập thành số nguyên
            try:
                so = int(chuoi_nhap)
                
                # 4. Ghi số đó vào file, CÓ KÈM theo ký tự xuống dòng '\n'
                f.write(f"{so}\n")
                
            except ValueError:
                # Xử lý nếu người dùng nhập không phải là số
                print("Lỗi: Đây không phải là số nguyên. Vui lòng nhập lại.")

    print(f"--- Đã ghi xong file '{ten_file}' ---")
    print("\n" + "="*30 + "\n") # In ra 1 dòng ngăn cách

    # --- PHẦN 2: ĐỌC FILE VÀ IN RA MÀN HÌNH ---
    print(f"--- Bắt đầu đọc file '{ten_file}' và in ra màn hình ---")
    
    # Mở file với chế độ 'r' (read - đọc)
    with open(ten_file, 'r', encoding='utf-8') as f:
        # f.read() sẽ đọc toàn bộ nội dung của file
        noi_dung_file = f.read()
    
    # In nội dung đã đọc
    if noi_dung_file:
        print("Nội dung file là:\n")
        print(noi_dung_file)
    else:
        print("File rỗng (bạn không nhập số nào).")

except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file '{ten_file}' để đọc.")
except IOError as e:
    print(f"Lỗi: Không thể ghi hoặc đọc file. Chi tiết: {e}")
```

```python
def fib(n):
    f = [0] * (n+1)
    f[1] = 1
    for i in range(2, n + 1):
        f[i] = f[i - 1] + f[i - 2]
    return f[n]
n = int(input())
print(fib(n))
```
