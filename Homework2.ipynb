```python
#Bai 3
a, b, c = map(float, input("Nhap 3 canh cua tam giac: ").split())

if(a + b > c and a + c > b and b + a > c):
    p = (a+b+c)/2
    print(f"Dien tich cua tam giac la: {(p*(p-a)*(p-b)*(p-c))**1/2:.1f}")
    print("Chu vi cua tam giac la:", p*2)
    if(a == b == c): 
        print("Deu")
    elif(a == b or b == c or c == a):
        print("Can")
    elif(a*a + b*b == c*c or a*a == b*b + c*c or a*a + c*c == b*b):
        print("Vuong")
else: print('Khong la ba canh cua tam giac')
```

```python
#Bai 4
a = float(input("Nhập hệ số a: "))
b = float(input("Nhập hệ số b: "))

print(f"Giải phương trình {a}x + {b} = 0")

if a == 0:
    if b == 0:
        print("Phương trình có vô số nghiệm.")
    else:
        print("Phương trình vô nghiệm.")
else:
    x = -b / a
    print(f"Phương trình có nghiệm duy nhất x = {x}")
```
